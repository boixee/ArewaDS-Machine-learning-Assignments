# Exploring Visualizations

### Create some visualizations using the Pumpkin data in this lesson with matplotlib and seaborn in a sample notebook.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import datasets
```

```python
pumpkin = pd.read_csv('data/US-pumpkins.csv')
pumpkin.head()
```

```python
pumpkin.info()
```

```python
plt.scatter('Item size', 'Low Price')
plt.xlabel('Low price')
plt.ylabel('Item size')
plt.title('pumpkin data set (Low price vs Item size)')
plt.grid(True)
plt.show()
```

```python
pumpkin.groupby(['Type'])['Low Price'].mean().plot(kind='bar')
plt.ylabel("Pumpkin Price")
plt.grid(True)
```

```python
sns.kdeplot(pumpkin)
plt.ylabel=('Density')
```

###  Which libraries are easier to work with?

Matplotlib is better for basic plots while seaborn is better for more advanced statistical plots. and matplotlib has a more extensive library of function than seaborn.
