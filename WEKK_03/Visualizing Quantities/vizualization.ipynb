# visualizing data about Snow Goose

We are going to visualize the Snow Goose data using Line chart, Bar chart and Scatter plot. So now we're going to imports our working Libraries with is pandas and matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

After we imported labrarie, lets also call our data by using pandas  

```python
birds = pd.read_csv('../Visualizing Quantities/data/birds.csv')
birds.head()
```

Great! Our dataset is ready, now lets deeper and work, but we will only focus on Snow Goose Birds

```python
Snow_goose = birds.loc[[2], :]   
Snow_goose.head()      
```

```python
Snow_goose.plot()
plt.title('MinLength vs MaxBodyMass')
plt.ylabel('MinLength')
plt.xlabel('MaxBodyMass')

x = ['MinLength']
y = ('MaxBodyLength')

plt.show()
```

```python
minLength = Snow_goose['MinLength']
maxLength = Snow_goose['MaxLength']


plt.bar(minLength, maxLength)
plt.title('minLength vs maxLength')
plt.rcParams['figure.figsize'] = [3,6]

plt.show()
```

```python
minLength = Snow_goose['MinLength']
maxLength = Snow_goose['MaxLength']


plt.scatter(minLength, maxLength)
plt.title('minLength vs maxLength')
plt.grid(True)

plt.show()
```
