# Visualizing Distributions

I will use a simple dataset to learn how histogram helps to understand a dataset. Now lets import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

I will use python to make the plots. Let’s import the dataset:

```python
df = pd.read_csv('../Visualizing Distributions/data/Cartwheeldata.csv')
df.head()
```

This dataset shows cartwheel data.  And the dataset above shows the results. Let’s understand the data.

1. Make a histogram of ‘Age’.

```python
df['Age'].plot(kind = 'hist', bins = 10, figsize = (6,6))
plt.title("Histogram of age")
plt.grid(True)
plt.show()
```

From the picture above, it is clear that most people are below 30. Only one person is 39and one is 54. The distribution is right-skewed.

2. Make a distribution of ‘CWDistance’.

```python
df['CWDistance'].plot(kind = 'hist')
plt.title("Histogram of CWDistance")
plt.grid(True)
plt.show()
```

 It is hard to say which range has the most frequency.

3. Lets Plot ‘Height’ and ‘CWDistance’ in the same figure.

```python
df['Height'].plot(kind = 'hist')
df['CWDistance'].plot(kind = 'hist')
plt.title("Histogram of height and score")
plt.grid(True)
plt.show()
```
