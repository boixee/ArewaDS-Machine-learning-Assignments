```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                         columns= iris['feature_names'] + ['target'])
```

```python
iris_data.head()
```

Scatterplot using Matplotlib 

```python
plt.scatter(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'], c=iris_data['target'])
plt.figure(figsize=(8, 6))
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('sepal length vs sepal width (cm)')
plt.show()
```

Scatterplot using Seaborn

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target',data=iris_data)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.title('petal length vs petal width (cm)')
plt.show()
```

The scatterplot above represent different features of the iris dataset and relationship. in the first Scatterplot above, we plot the sepal length against sepal width,with each data point colored by its target class.

In the second scatterplot, we plot the petal length against the petal width each data point colored by its target class.
